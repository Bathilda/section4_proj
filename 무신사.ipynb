```python
def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    cloth_indices = [i[0] for i in sim_scores]
    return titles.iloc[cloth_indices]
```

```python
# 인코딩을 위한 전처리 진행

user_ids = df["label_id"].unique().tolist()
user2user_encoded = {x: i for i, x in enumerate(user_ids)}
userencoded2user = {i: x for i, x in enumerate(user_ids)}
mu_ids = df["label_title"].unique().tolist()
mu2mu_encoded = {x: i for i, x in enumerate(mu_ids)}
mu_encoded2mu = {i: x for i, x in enumerate(mu_ids)}
df["user"] = df["label_id"].map(user2user_encoded)
df["mu"] = df["label_title"].map(mu2mu_encoded)

num_users = len(user2user_encoded)
num_mu = len(mu_encoded2mu)
df["rating"] = df["rating"].values.astype(np.float32)
# min and max ratings will be used to normalize the ratings later
min_rating = min(df["rating"])
max_rating = max(df["rating"])

print(
    "Number of users: {}, Number of Movies: {}, Min rating: {}, Max rating: {}".format(
        num_users, num_mu, min_rating, max_rating
    )
)
```

```python
EMBEDDING_SIZE = 100


class RecommenderNet(keras.Model):
    def __init__(self, num_users, num_mu, embedding_size, **kwargs):
        super().__init__(**kwargs)
        self.num_users = num_users
        self.num_mu = num_mu
        self.embedding_size = embedding_size
        self.user_embedding = layers.Embedding(
            num_users,
            embedding_size,
            embeddings_initializer="he_normal",
            embeddings_regularizer=keras.regularizers.l2(1e-6),
        )
        self.user_bias = layers.Embedding(num_users, 1)
        self.mu_embedding = layers.Embedding(
            num_mu,
            embedding_size,
            embeddings_initializer="he_normal",
            embeddings_regularizer=keras.regularizers.l2(1e-6),
        )
        self.mu_bias = layers.Embedding(num_mu, 1)

    def call(self, inputs):
        user_vector = self.user_embedding(inputs[:, 0])
        user_bias = self.user_bias(inputs[:, 0])
        mu_vector = self.mu_embedding(inputs[:, 1])
        mu_bias = self.mu_bias(inputs[:, 1])
        dot_user_mu = tf.tensordot(user_vector, mu_vector, 2)
        # Add all the components (including bias)
        x = dot_user_mu + user_bias + mu_bias
        # The sigmoid activation forces the rating to between 0 and 1
        return tf.nn.sigmoid(x)


model = RecommenderNet(num_users, num_mu, EMBEDDING_SIZE)
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
)
```

```python
history = model.fit(
    x=x_train,
    y=y_train,
    batch_size=128,
    epochs=10,
    verbose=1,
    validation_data=(x_val, y_val),
)
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["train", "test"], loc="upper left")
plt.show()
```
